using translate and maketrans : 
https://www.w3schools.com/python/ref_string_maketrans.asp

```python
# DELETING PUNCTUATION
import re, string

def deletepunctuation(s):
  s = s.strip()
  res = s.translate(str.maketrans("\'—-–", "    ", "!\"$%&()*+,./:;<=>?_@[\]^`{|}~"))
  res = res.strip()
  return res

df["context"] = df["context"].apply(deletepunctuation)
df["question"] = df["question"].apply(deletepunctuation) #??
df["text"] = df["text"].apply(deletepunctuation)

df.head()
```

##ADDING NEW COLUMNS
STRING-TO-LIST-OF-WORDS COLUMNS AND THE ONES FOR THE INDEXES OF ANSWER-START AND ANSWER-END

**QUESTION_LIST: come si chiama il padre di Lucia?**

**CONTEXT_LIST: il(0) padre(1) di(2) Lucia(3) si(4) chiama(5) Tizio(6) Caio(7)**

**ANSWER_LIST: Tizio(0) Caio(1)**

to find the **match** with the ANSWER_LIST "Tizio Caio",we check the possible matches between the elements of ANSWER_LIST and some of CONTEXT_LIST, in particular those from **index+len(ANSWER_LIST)** for all the CONTEXT_LIST elements

```python
# FINDING ANSWER_START AND ANSWER_END
def find_sublist(context_list, answer_list):
  for i, word in enumerate(context_list):
    if context_list[i:i+len(answer_list)]==answer_list:
      return (i, i+len(answer_list)-1)
  return (-1, -1)
```

```python
# CREATING THE LISTS FROM THE STRINGS IN THE DF + FINDING THE INDEXES
rows = []

for i, row in df.iterrows():
  question_list = row["question"].split(" ") 
  context_list = row["context"].split(" ")
  answer_list = row["text"].split(" ")
  start, end = find_sublist(context_list, answer_list)
  
  new_cols = {
      "context_list" : context_list,
      "question_list" : question_list,
      "answer_list" : answer_list,
      "answer_start" : start,
      "answer_end" : end
  }

  rows.append(new_cols)
```

```python
#ADDING THE NEW COLUMNS
df[["context_list", "question_list", "answer_list", "answer_start", "answer_end"]] = pd.DataFrame(rows)
```

```python
def clean_string(s):
  s = str(s)
  s = s.lower()
  s = s.strip()
  res = s.translate(str.maketrans("\'—-–", "    ", "!\"$%&()*+,./:;<=>?_@[\]^`{|}~"))
  res = res.strip()
  return res
```
