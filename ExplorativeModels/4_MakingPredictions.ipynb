```python
# system packages
import os
import shutil
import sys

# data and numerical management packages
import pandas as pd
import numpy as np

# useful during debugging (progress bars)
from tqdm import tqdm


from urllib import request
import json
```

##IMPORTING VOCABULARIES

```python
glove_idx_to_word = {}
glove_word_to_idx = {}
trainable_idx_to_word = {}
trainable_word_to_idx = {}

with open('./NLP/vocabularies/1_2_3_initialmodels/glove/idx_to_word.json') as f:
    glove_idx_to_word = json.load(f)
with open('./NLP/vocabularies/1_2_3_initialmodels/glove/word_to_idx.json') as f:
    glove_word_to_idx = json.load(f)
with open('./NLP/vocabularies/1_2_3_initialmodels/trainable/idx_to_word.json') as f:
    trainable_idx_to_word = json.load(f)
with open('./NLP/vocabularies/1_2_3_initialmodels/trainable/word_to_idx.json') as f:
    trainable_word_to_idx = json.load(f)
```

##IMPORTING TEST SET

```python
# DATAFRAME ROWS
dataframerows = []

with open("./NLP/FINAL_DATASETS/VERSION_IDX_BY_WORD/test_set.json") as f:
    data = json.load(f)

    for el in data['data']:
        title = el['title']
        paragraphs = el['paragraphs']
        
        for context_qas in paragraphs:
              context = context_qas['context'] 
              qas = context_qas['qas']

              for a_q in qas: #[0]["answers"][0]["text"]
                  answer = a_q['answers']
                  question = a_q['question']
                  id = a_q['id']

                  # CREATE ROW
                  row = {
                      "id" : id,
                      "context" : context,
                      "question" : question,
                      }
                  
                  # APPEND ROW TO DATAFRAME ROWS
                  dataframerows.append(row)

df_test = pd.DataFrame(dataframerows)
df_test = df_test[["id", "context", "question"]]
```

```python
df_test.head()
```

##SPLITTING BOTH CONTEXT AND QUESTIONS IN LISTS

```python
# CREATING THE LISTS FROM THE STRINGS IN THE DF TO LATER US THEM WITH THE INDEXES PREDICTED BY THE MODELS
rows = []

for i, row in df_test.iterrows():
  question_list = row["question"].split(" ") 
  context_list = row["context"].split(" ")
  
  new_cols = {
      "context_list" : context_list,
      "question_list" : question_list,
  }

  rows.append(new_cols)
```

```python
#ADDING THE NEW COLUMNS
df_test[["context_list", "question_list"]] = pd.DataFrame(rows)
```

```python
df_test.head()
```

##CHECKING IF THERE ARE OUT-OF-VOCABULARY WORDS

THE VOCABULARY IS CREATED ACCORDING TO TRAIN AND VALIDATION SETS

```python
from pandas.core.common import flatten

corpus = list(flatten(df_test['question_list']))
corpus += list(flatten(df_test['context_list']))
corpus = set(corpus)
glove_oovterms = [term for term in corpus if term not in list(glove_word_to_idx.keys())]
trainable_oovterms = [term for term in corpus if term not in list(trainable_word_to_idx.keys())]

print("GLOVE - THERE ARE {} OUT-OF-VOCABULARY TERMS: {:.1f}% OF THE TEST SET'S TERMS".format(len(glove_oovterms), len(glove_oovterms)/len(corpus)*100))
print("TRAINABLE - THERE ARE {} OUT-OF-VOCABULARY TERMS: {:.1f}% OF THE TEST SET'S TERMS".format(len(trainable_oovterms), len(trainable_oovterms)/len(corpus)*100))
```

THEY SHOULD BE EXACTLY THE SAME VOCABULARIES,

AND IN FACT WE GET THEY MISS THE SAME NUMBER OF THE TEST-SET'S WORDS,

BUT WE USE TWO JUST TO MAKE SURE WE WILL NOT HAVE PROBLEMS WITH INDEXES ACCORDANCE WITH THE EMBEDDING MATRICES.

#PREDICTIONS

##PREDICTING WITH THE GLOVE MODELS

###ENCODING TEST SET

```python
import tensorflow as tf

# ENCODING OF THE TEST SET (WHOLE SET AT ONE TIME)
    # CONTEXT_LIST => EACH ELEMENT ENCODED AS ITS VOCABULARY'S INDEX
    # QUESTION_LIST => EACH ELEMENT ENCODED AS ITS VOCABULARY'S INDEX

def encode(df, idx_to_word, word_to_idx, oov):
    contexts = []
    questions  = []
    
    for i, row in df.iterrows():
    #COLUMN NAMES : question_list, context_list
        # QUESTIONS EMBEDDING
        question = []
        # formatting single question
        for word in row['question_list']:
          if word in oov:
            question.append(word_to_idx['OOV'])
          else:
            question.append(word_to_idx[word])
        # adding it to the questions' column
        questions.append(question)
        
        # CONTEXTS EMBEDDING
        context = []
        # formatting single context 
        for word in row['context_list']:
          if word in oov:
            context.append(word_to_idx['OOV'])
          else:
            context.append(word_to_idx[word])
        # adding it to the contexts' column
        contexts.append(context)
    
    contexts = tf.keras.preprocessing.sequence.pad_sequences(contexts, dtype="float32", padding="post")
    questions = tf.keras.preprocessing.sequence.pad_sequences(questions, dtype="float32", padding="post")
    
    return (questions, contexts)
```

```python
df_testENCglove = encode(df_test, glove_idx_to_word, glove_word_to_idx, glove_oovterms)
```

###GLOVE MODEL1
- ATTENTION LAYER
- REDUCIBLE LEARNING RATE (0.01-0.001)
- BATCH SIZE 256

This is also called our "Standard Model" because it was the first over which we made all the alternatives and changes in order to find possible improvements

####PREDICT

```python
from tensorflow.keras.models import load_model

# Path where the model is saved
glove_model1_attention = os.path.join("model1_attention.h5")

# Uploading the weights
glove_model1 = load_model(glove_model1_attention)
```

```python
glove_pred1 = glove_model1.predict(df_testENCglove, verbose=0)
glove_pred1 = list(zip(glove_pred1[0], glove_pred1[1]))
glove_predidxs1 = np.argmax(glove_pred1, axis=2)
```

```python
well_formed = 0
for el in glove_predidxs1:
  if el[0]<=el[1]:
    well_formed += 1

print("WELL FORMED PREDICTIONS ARE {}, {:.1f}%".format(well_formed, well_formed/len(df_test)*100))
print("EMPTY ANSWERS WILL BE {}, {:.1f}%".format(len(df_test)-well_formed, 100-well_formed/len(df_test)*100))
```

####ANSWER EXTRACTION
FROM THE PREDICTED INDEXES TO THE CORRESPONDENT SLICE OF CONTEXT LIST

FROM SLICE OF LIST TO SINGLE STRING

```python
glove_finalpred1 = {}

for i in range(len(df_test)):
    answer = df_test['context_list'][i][glove_predidxs1[i][0]: glove_predidxs1[i][1]+1]  
    glove_finalpred1[df_test['id'][i]] = ' '.join(answer)
```

```python
#SHOWING SOME EXAMPLES
for key in list(glove_finalpred1.keys())[:5]:
    print(key, glove_finalpred1[key])
```

####SAVING PREDICTED ANSWER IN A JSON FILE

```python
# convert into JSON:
with open('glove_predictions_model1_attention.json', 'w') as outfile:
  json.dump(glove_finalpred1, outfile)
```
